## 2. The question: decision, action, cost of a wrong call

*What decision does your work improve? Who acts on it? What does a wrong recommendation cost?*

### Decision
Which pages should an SEO reviewer prioritize for content refresh?

### Unit of analysis
One content page.

### Output
A ranked priority score representing the likelihood that a page deserves review.

### Action
The SEO team reviews the highest-ranked pages first and decides whether to refresh, expand, rewrite, protect, or monitor the content.

### Cost of a wrong recommendation
False positives waste reviewer time by prioritizing pages that did not require action.

False negatives are more expensive because genuinely valuable pages may continue losing visibility or traffic while being ignored.

### Why data and ML?
Simple rule-based scoring cannot capture complex interactions between visibility, freshness, engagement, CTR, and search trends. A machine learning model may learn these relationships and produce a better prioritization than fixed rules while still supporting human decision-making.

## 4. What I can and can't claim

This project aims to support content prioritization using observed search and engagement signals. Any rankings produced by the model should be interpreted as decision-support rather than proof that a page requires a refresh.

The analysis can identify patterns associated with pages that may deserve review, but it cannot explain why Google ranked a page a certain way or prove that refreshing content will improve search performance. Any recommendations should therefore be treated as evidence-based suggestions that still require human review.
