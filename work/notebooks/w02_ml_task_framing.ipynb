## 1. My lane as an ML task (type)

*Classification, clustering, ranking, or scoring — which one, and why?*

I chose **Refresh / Content Opportunity Scoring**.

The underlying ML task is **classification**. For each content page, the model estimates whether it is likely to belong to the "needs review" group based on observed search and engagement signals.

Although the final output is a ranked review queue, the ranking is produced from predicted probabilities generated by a classification model. Those probabilities help prioritize which pages should be reviewed first.

## 2. Target or proxy

*What would you predict? Where does that label come from — observed outcome or a defined rule?*

For the starter dataset, I will use the provided proxy label:

**is_declining_label = (trend_direction == "down")**

This is a rule-based proxy rather than a true future outcome. It represents pages whose observed search trend is currently declining.

The lane guide explains that this is suitable for the starter dataset but that a stronger capstone would define a future-looking label, such as using the previous 90 days of data to predict decline over the next 30 days.

## 3. Success metric

*One metric you can defend. What number means 'good'?*

The primary success metric is **Precision@50**.

This metric measures how many of the top 50 pages recommended by the model are actually positive according to the chosen label.

Precision@50 matches the real business decision because SEO teams usually review only a limited number of pages at a time. A higher Precision@50 means reviewers spend more time on pages that genuinely deserve attention.

## 4. The unit of analysis, as a real dataframe

*Load your lane's slice and show it: one row = one what?*

```python
import pandas as pd

df = pd.read_csv("content_refresh_anonymized.csv")

print("Dataset shape:", df.shape)

df["is_declining_label"] = (df["trend_direction"] == "down").astype(int)

lane_df = df[
    [
        "content_id",
        "impressions_90d",
        "sessions_90d",
        "ctr",
        "avg_position",
        "trend_direction",
        "is_declining_label",
    ]
]

lane_df.head(10)
```

## Unit of analysis

One row represents **one content page**.

Each row contains observed search and engagement measurements for a single page over the selected time window, including impressions, sessions, CTR, average search position, and trend information. The target column (`is_declining_label`) is a proxy label derived from the observed `trend_direction`, where a value of 1 indicates a declining page and 0 indicates a stable or improving page.

This page-level representation matches the business decision because content reviewers evaluate and prioritize individual pages for refresh or optimization.

## 5. Why ML beats a fixed rule here

*What makes the pattern too messy for an if-statement?*


A fixed rule might identify pages using simple conditions such as "impressions > 500" or "CTR < 0.5". While these rules are easy to understand, they cannot capture how multiple signals interact.

For example, a page with high impressions but strong CTR may not require attention, while another page with moderate impressions, declining trends, low engagement, and aging content could be a much stronger candidate for review. These combinations are difficult to represent with a small set of if-else statements.

A machine learning model can learn patterns from several observed features simultaneously, such as impressions, sessions, CTR, average position, freshness, and engagement. Instead of replacing human judgment, it assigns a probability that helps rank pages for review.

The goal is not to automate content decisions but to provide decision support by prioritizing pages that are more likely to benefit from human review.
